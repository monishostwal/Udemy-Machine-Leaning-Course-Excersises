```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
xe=np.random.uniform(low=-10,high=10,size=(1000,1))
ze=np.random.uniform(low=-10,high=10,size=(1000,1))
```

```python
inputs=np.column_stack((xe,ze))
```

```python
inputs.shape
```

```python
noise=np.random.uniform(-1,1,(1000,1))
targets=2*xe+3*ze+5+noise
```

```python
targets.shape
```

```python
np.savez("./Prepossesed Dataset/bnn",inputs=inputs,targets=targets)
```

```python
loaded_data=np.load('./Prepossesed Dataset/bnn.npz')
```

```python
loaded_data['inputs']
```

```python
input_size=2
output_size=1
model=tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(loaded_data['inputs'],loaded_data['targets'],epochs=100,verbose=1)
```

```python
model.layers[0].get_weights()
```

```python
model.predict_on_batch(loaded_data['inputs'])
```

```python
plt.plot(model.predict_on_batch(loaded_data['inputs']),loaded_data['targets'])
plt.show()
```
