```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv('./Datasets/countries with continent.csv')
data
```

```python
data['continent'].values
```

```python
new_Data=data.copy()
```

```python
new_Data['continent'] = new_Data['continent'].map({'North America':0,'Europe':1,'Asia':2,'Africa':3,'South America':4, 'Oceania':5,'Seven seas (open ocean)':6, 'Antarctica':7})
new_Data
```

```python
x=new_Data.iloc[:,1:4]
```

```python
x
```

```python
kmeans=KMeans(4)
```

```python
kmeans.fit(x)
```

```python
predicted_res=kmeans.fit_predict(x)
```

```python
predicted_res
```

```python
final_data=x.copy()
```

```python
final_data['cluster']=predicted_res
```

```python
final_data
```

```python
plt.scatter(final_data['Latitude'],final_data['Longitude'],c=final_data['cluster'],cmap='rainbow')
plt.show()
```
